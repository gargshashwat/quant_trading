```python
import matplotlib.pyplot as plt
import sys
sys.path.append('./venv/lib/python3.7/site-packages')
import pymysql

from dummy_alpha import *
from utils import *
from portfolio import *
from stats_report import *

%load_ext autoreload
%autoreload 2
```

```python
start = '2015-01-01'
end = '2021-01-01'

N = get_df_from_db(f"""SELECT * FROM mapping""").id.tolist()[:500]
id_string = '(' + ','.join([f"'{i}'" for i in N]) + ')'

df = get_df_from_db(f"""SELECT id, date, close FROM price_volume_US
                            JOIN mapping ON price_volume_US.ticker=mapping.ticker
                            WHERE price_volume_US.date >= {pd.to_datetime(start).strftime('%Y%m%d')}
                            AND price_volume_US.date <= {pd.to_datetime(end).strftime('%Y%m%d')}
                            AND id in {id_string}""")

# pivot
df = df.pivot(index='date', columns='id', values='close')

# filter

# fill nan values
df = df.fillna(method='ffill', axis=0).fillna(0)

# convert to numpy array
close = np.array(df)
```

```python
alpha = Alpha()
alpha.N = df.columns
alpha.T = df.index
alpha.region = 'US'
alpha.name = 'momentum'
        
mom_close = momentum(close, 5, 60)
alpha.alpha = shift(mom_close, 20)  # delay alpha by one day

alpha.alpha = normInv(alpha.alpha)
```

```python
ret = get_returns(alpha.N, alpha.T)
```

```python
alpha.raw_ic_analysis()
```

```python
pnls, _ = calculate_pnl(alpha.alpha, ret, alpha.T)
irs, _ = calculate_ir(pnls)
irs
```

```python
w = characteristic_portfolio(alpha)
```

```python
alpha2 = Alpha()
alpha2.N = alpha.N
alpha2.T = alpha.T
alpha2.region = 'US'
alpha2.name = 'momentum holdings'

alpha2.alpha = w
```
