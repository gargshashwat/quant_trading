```python
start = '2010-01-01'
end = '2021-01-01'

N = get_df_from_db(f"""SELECT * FROM mapping""").id.tolist()[:500]
id_string = '(' + ','.join([f"'{i}'" for i in N]) + ')'

df = get_df_from_db(f"""SELECT id, date, close FROM price_volume_US
                            JOIN mapping ON price_volume_US.ticker=mapping.ticker
                            WHERE price_volume_US.date >= {pd.to_datetime(start).strftime('%Y%m%d')}
                            AND price_volume_US.date <= {pd.to_datetime(end).strftime('%Y%m%d')}
                            AND id in {id_string}""")

# pivot
df = df.pivot(index='date', columns='id', values='close')

# fill nan values
df = df.fillna(method='ffill', axis=0).fillna(0)

# convert to numpy array
close = np.array(df)
```

```python
def std_rolling_ts(arr, window):
    in1 = pd.DataFrame(arr).fillna(0)
    std_ts = np.array(in1.rolling(window).std())
    return std_ts
    

def momentum(data, hl_short, hl_long):
    fast = ewma(data, hl_short, 2)
    slow = ewma(data, hl_long, 2)
    diff = fast - slow
#     diff /= diff.std(axis=1).reshape(-1,1)
    diff /= std_rolling_ts(diff, 60)
    return diff.clip(-3,3)

def momentum2(arr, hl):
    in1 = shift(arr, 40)
    in1 = np.array(pd.DataFrame(in1).fillna(0).rolling(120).mean().fillna(0))
    return normInv(ewma(in1, hl))

def dewma(arr, hl, nhl):
    ewma1 = ewma(arr, hl, nhl)
    ewma2 = ewma(ewma1, hl, nhl)
    return ewma1  + (ewma1 - ewma2)

def daily_to_monthly(arr):
    arrs = []
    for i in np.arange(20):
        arr_i = arr.copy()
        idx = [t for t in np.arange(arr.shape[0]) if t%20 != i]
        arr_i[idx] = np.nan
        arr_i = np.array(pd.DataFrame(arr_i).fillna(axis=0, method='ffill', limit=20))
        arrs.append(arr_i)
    return np.array(arrs).mean(axis=0)


def daily_to_monthly2(arr, d):
    arrs = []
    for i in np.arange(20):
        arr_i = arr.copy()
        idx = [t for t in np.arange(arr.shape[0]) if t%20 != i]
        arr_i[idx] = np.nan
        arr_i = np.array(pd.DataFrame(arr_i).fillna(axis=0, method='ffill', limit=20))
        arrs.append(arr_i)
    return arrs[d]
```
